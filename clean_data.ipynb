Trails Data:

```python
trails_data_to_load = "./Resources/trail_data.csv"
trails_df = pd.read_csv(trails_data_to_load, low_memory=False, encoding="utf-8")
trails_df.head(5)
```

```python
renamed_trails_df = trails_df.rename(columns={"area_name": "park_name",
                                            })
renamed_trails_df.head()
```

```python
trail_count = trails_df.groupby(["area_name"]).count()["trail_id"]
print(trail_count)
```

```python
trail_pop_avg = trails_df.groupby(["area_name"]).mean()["popularity"]
print(trail_pop_avg)
```

```python
# Assign empty list for park names
park_name_options = []

with open(trails_data_to_load) as trails_data:

    # To do: perform and read analysis data.
    reader = csv.reader(trails_data)

    #Read the header row
    #header = next(file_reader)

    #Iterate each row in the trail_data.csv file.
    for row in reader:
        
        # Print the park name from each row
        park_name = row[2]
        
        # If the park name does not match any existing candidate add it to park name options list
        if park_name not in park_name_options:
            
            #Add it to the list 
            park_name_options.append(park_name)
            
park_name_options
```

Biodiversity Data

```python
biodiversity_data_to_load = "./Resources/species.csv"
bio_df = pd.read_csv(biodiversity_data_to_load, low_memory=False)
bio_df.head()
```

```python
bio_count = bio_df.groupby(["Park Name"]).count()["Species ID"]
print(bio_count)
```
